# Importing Required Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Loading DataSet to a Pandas DataFrame

```python
df= pd.read_csv('/creditcard.csv')
df.head()
```

```python
# reading last 5 rows from the dataframe
df.tail()
```

```python
#Infromation about dataset
df.info()
```

Checking Missing values

```python
df.isnull().sum()
```

we can observe that there are no missing values and hence we can continue.

```python
df['Class'].value_counts()
```

distribute transactions on the basis of class
for Class=0 we say those are normal transactions i.e., legit transactions (120651)
whereas Class=1 are fradulent transactions(249)

This dataset is highly Unbalanced.

```python
#seperating data for analysis.
legit= df[df.Class== 0]
fraud=df[df.Class==  1]
```

```python
print(legit.shape)
print(fraud.shape)
```

statistical measures of data

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

comparing the values for both transactions (based on mean value)

```python
df.groupby('Class').mean()
```

Under-Sampling

Build a sample dataset having similar distribution of Legit and Fraudulent transactions

number of Fraudulent Transactions=249

therefore add random 249 transactions from normal transactions and add it to Fraudulent transactions so as to make uniform distribution of data in both the cases

```python
legit_sample= legit.sample(n=249)
```

Concatinating Two DataFrames.

```python
new_df= pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df['Class'].value_counts()
```

uniformly distributed data is formed with 249 normal and 249 Fraudulent transactions.

```python
new_df.groupby('Class').mean()
```

We can observe that there is not much difference in mean values before and after adding transaction values into fraudulent transactions.(i.e., nature of dataset is not changed which helps our machine for easy detection of fraud transactions.)

Splitting the data into Features and Targets(Class)

```python
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
print(X)
```

```python
print(Y)
```

**Applying Train_Test Split method**

split the data into testing data(20%) and training data(80%)
 and stratifying data according to 'Y'

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.2, stratify=Y,random_state=2)
```

```python
#checking sizes of dataset
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

**Logistic Regression**

```python
model= LogisticRegression()
```

training logistic regression model with training data

```python
model.fit(X_train, Y_train)
```

Model Evaluation

Finding Accuracy (accuracy score function)

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy =  accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data', training_data_accuracy)
```

```python
X_test_prediction =  model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on test data', test_data_accuracy)
```

**CONCLUSION**

 As the DataSet is distributed uniformly now, and

 accuracy score of both training data and testing data is almost similar i.e., data set is not overfitted and underfitted

 therefore, credit card fraud detection will be done more accurately by machine.
